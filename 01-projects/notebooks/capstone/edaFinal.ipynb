Georgia's public schools are evaluated and held accountable for their policies with a metric called the College and Career Ready Performance Index, a score that is calculated annually for each public school in the state, and the Department of Education describes it as the best measurement to ensure that schools are "preparing students for the next educational level." 

The score is determined by a series of indicators, nearly all of which are calculated based on student performance on state assessments. Schools are then compared to other schools across the state, their couny, and their respective school districts.

While having a measure to hold schools accountable and share best practices between schools is essential, comparing schools only makes sense if the schools are grouped according to the makeup of the students they serve. The classic clustering method for public schools is to group them by districts, yet there's extreme disparity between the demographics of schools within a district, making it difficult for struggling schools to learn and implement best practices from schools making progress despite similar circumstances.

In regards to accountability, the CCRPI score gives scores a "Report Card," which, similar to a student's report card, transfers their CCRPI score to a letter grade. However, this metric, again, only compares a school to all others. This would be more effective if there were a way to compare a school's grade to what their grade *should* be, based on what schools with similar student populations have scored.

By collecting the data that's described in greater detail below, I'll attempt to reorganize Georgia's schools into groups that more accurately reflect similarities in their student populations with various clustering algorithms. I'll then process the data with various classification methods in order to predict what a school's CCRPI should be, so it can be compared to their actual score.

Using the dashboard below, you can filter through any school district in the state, or choose multiple districts if you wish. Schools are automatically divided by the grades they serve, and there's a slider to choose a range of scores. This dashboard can assist in better understanding the issue of comparing schools based on traditional organization.
 
The basic flaw in the CCRPI score is an organization problem. The fundamental unit in which schools are organized is their respective district, and to a larger extent, their counties. While these units may have once (but probably not) consisted of relatively homogenous student populations, that's no longer the case. Two schools may share a district, and will then be compared to one another with the same evaluative metric. However, this poses a problem, as two schools may serve student populations of vastly different sizes, backgrounds, needs, and may have very different levels of staff experience or student-to-teacher ratio.

For example, if you select Atlanta Public Schools as the district, you can immediately see how widely dispersed these schools are across the city. There's also a great disparity between those schools in terms of the students they teach and the operations under which they operate. For example, the lowest performing schools in 2015 were Intown Charter Academy and Harper-Archer Middle School. Their performance is then compared to other middle schools in their district, the top performers of which were Inman Park Middle and Kindezi Academy. However, Kindezi Academy, for example, is a charter school whose rules allow them to have extremely small class sizes, for example, while a traditional public school like Harper-Archer is bound to larger class sizes. If a small student-teacher ratio positively impacts CCRPI performance, it doesn't make sense to then compare the two schools as if they're operating in exact circumstances. More importantly, if the overall goal is to improve the performance of students at Harper Archer - as it should be - they can't learn much from a school like Kindezi, since Harper-Archer isn't able to simulate those incredibly small class sizes, which account for the foundation of Kindezi's academic culture.

```python
HTML('<iframe width="1200" height="700" src=https://public.tableau.com/views/CCRPICorrelations2/StudentTeacherRatio?:showVizHome=no&:embed=true frameborder="0" allowfullscreen=true></iframe>')
```

Surprisingly, student-teacher ratio doesn't seem to have much impact on a school's CCRPI performance. In fact, the trends suggest a higher student-teacher ratio increases scores. Nevertheless, if student-teacher ratio were to matter, it makes no sense to compare Harper's performance to Kindezi's performance, as they're on opposite ends of the spectrum.
 
Instead, we can take a school metric, like its student-teacher ratio, and split schools into groups based on similar student-teacher ratios, and then compare them to schools in their respective distributions. Notice that Kindezi and Harper-Archer are placed in very different buckets. Now things are set equal for the schools. Kindezi doesn't have a peer (except for its own elementary school), while Harper has many, and is performing worse than all of them. Also, by comparing groups based on a measurement like this, school leaders at Harper - or any other school - can now find peers across the entire state, as they can simply filter through whichever districts they prefer, and share best practices with schools outside their own immediate zone.

Let's look at another example with a more influential metric: the percentage of a school's students whose families receive TANF or SNAP benefits, and/or are homeless.

```python
HTML('<iframe width="1200" height="700" src=https://public.tableau.com/shared/Y8MTKY82G?:showVizHome=no&:embed=true frameborder="0" allowfullscreen=true></iframe>')
```

Here, the relationship between the size of a school's student population coming from economically struggling households and that school's CCPRI score is negative. This makes intuitive sense and has been supported by research for decades. Children that are economically disadvantaged generally perform worse in schools compared to their more affluent peers, as they have access to fewer resources and experiences that can encourage, expand, or reward learning, among many other differences.
 
Going back to our comparison of two schools: Harper-Archer Middle and Kindezi Academy. Despite sharing a school district, they serve student populations from very different socio-economic backgrounds. Comparing them to one another isn't very productive, but by placing them in a distribution with their actual peers, both schools can compare their performance to similar middle schools, and they don't have to be limited to schools in their district. Now, Kindezi is still performing better than most of its peers, but there are schools outperforming Kindezi, so its school leaders could perhaps learn some new practices that will improve their own students' learning outcomes even further.
 
Harper-Archer is still a low-performing school, but at least it can now look to schools with a similar student population and learn some useful practices. This is the basis of my hypothesis - if we could just regroup schools more accurately, it would be easier to compare apples to apples, and therefore make the CCRPI a more useful tool.
 
However, it's more complicated than simply regrouping schools based on their students' socio economic status. There are hundreds of factors that make a student population what it is, so we need to somehow find a combination of the relevant factors that truly separate school groups. For example, below are just a handful of school population metrics and their relationship with CCRPI scores.

```python
#Normalize the feature columns so the x axes in our subplots are all on the same scale.
cols_to_norm = ['stdTchRatio',
          'expPerStudent','salBenSum','freeLnch','reducedLnch','pctMobile','pctAbs5orLess','pctAbsSix-Fiftn',
          'pctOverFiftn','pctSnapTanfHmls', 'pctAsian','pctNative','pctBlck','pctHisp','pctMulti','pctWhite',
         'pctMigrant','pctSPED','pctESOL','pctAltProg','pctGiftd']
df2[cols_to_norm] = df2[cols_to_norm].apply(lambda x: (x - x.mean()) / (x.max() - x.min()))
```

All of these features make it clear that some combination of them have an impact on CCRPI, and that combination is what we need to find in order to reorganize these schools.
 
Since the schools aren't in these hypothetical groups, we don't have any classification algorithms we can run, so instead we'll need to create the groups with an unsupervised clustering model. However, we first need to figure out which of our metrics on schools actually differentiate them. To do this, a simple scatter plot matrix like those above for each feature will show which have actual trends with CCRPI scores, and I'll feed those features to the clustering model.
 
Out of more than 50 metrics, those with the sharpest trends I noticed were whether or not a school was Title I, the percentage of absent students, the percentage of students receiving TANF or SNAP assistance, the percentage of students that were black or Asian, as well as the percentage of students that were gifted or 'mobile,' which again means a student enrolled late in the school year, or transferred/dropped out later in the school year.
